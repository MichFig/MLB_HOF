```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# read classic inductees file 
classic = "../Resources/Hall_of_Fame.csv"
classic_df = pd.read_csv(classic)
classic_df
```

```python
# read contemporary inductees file
contemp = "../Resources/hall_of_fame.2.csv"
contemp_df = pd.read_csv(contemp)
contemp_df
```

```python
# read master player file
player = "../Resources/master_player.csv"
player_df = pd.read_csv(player)
player_df
```

```python
batting = "../Resources/Batting.csv"
batting_df = pd.read_csv(batting)
batting_df
```

```python
nickname = "../Resources/nick_names.csv"
nickname_df = pd.read_csv(nickname)
nickname_df
```

```python
# split Player column for merging
new = nickname_df["Player"].str.split(" ", n = 1, expand = True) 
nickname_df["Name"]= new[0]
nickname_df["Last Name"]= new[1]  
nickname_df.drop(columns =["Player"], inplace = True) 
new
```

```python
# merge new and nickname for future merge
tag = nickname_df.append(new)
tag.head()
```

```python
# delete unneeded columns
tag_df = tag.drop(tag.columns[[0, 1, 6, ]], axis=1)
tag.head()
```

```python
# rename columns to prepare for merging
re_name_df = contemp_df.rename(columns = {'Column1': 'last', 'Column3': 'playerID', 'Column4': 'birth', 'Column5': 'death'}, inplace = False)
re_name_df
```

```python
# merging classic and contemporary dataframes
df_merge = pd.merge(classic_df, re_name_df, on='playerID')
df_merge.head()
```

```python
# remove duplicate and unneeded columns
df_merge = df_merge.drop(df_merge.columns[[2, 3, 4, 5, 6, 7, 12]], axis=1)
df_merge
```

```python
# merge file with player file
players = pd.merge(df_merge, player_df, on='playerID')
players
```

```python
# merge player file with tag
hof_df = pd.merge(players, tag_df, on='Name')
hof_df.head()


```
