```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

# Player table

```python
csv_file = "Resources/master_player.csv"
player_df = pd.read_csv(csv_file)
player_df.head()
```

# Hall of Fame table

```python
csv_file = "Resources/Hall_of_Fame.csv"
hof_df = pd.read_csv(csv_file)
hof_df.head()
```

### Clean up the hof table to only include inductee's 

```python
# Only include playerID, yearid, category where inducted = Y
# Cleaned up version will go into hof_inductees table

new_hof_df = hof_df[['playerID','yearid','category']].where(hof_df['inducted']=='Y').dropna()
new_hof_df.head()
```

# Batting Stats table

```python
csv_file = "Resources/Batting.csv"
batting_df = pd.read_csv(csv_file)
batting_df.head()
```

```python
# Only include primary stats of hof inductees
# Get a list of hof playerID's
hof_player_list = new_hof_df["playerID"].tolist()

# New Batting dataframe with only hof inductees
batting2_df = batting_df[['playerID','yearID','teamID','G','AB','R','H','2B','3B','HR','RBI','SB']]
boolean_series = batting2_df.playerID.isin(hof_player_list)
hof_batting_df = batting2_df[boolean_series]
hof_batting_df.head()
```

```python
# Combine all batters stats to generate career stats
#ruth_df = hof_batting_df.where(hof_batting_df['playerID']=='ruthba01').dropna()
#career_batting_df = ruth_df[['playerID','G','AB','R','H','2B','3B','HR','RBI','SB']].groupby(['playerID']).sum()
career_batting_df = hof_batting_df[['playerID','G','AB','R','H','2B','3B','HR','RBI','SB']].groupby(['playerID']).sum().reset_index()

career_batting_df
```

# Pitching Stats table

```python
csv_file = "Resources/Pitching.csv"
pitching_df = pd.read_csv(csv_file)
pitching_df.head()
```

```python
# New Pitching dataframe with only hof inductees
pitching2_df = pitching_df[['playerID','yearID','teamID','W','L','GS','CG','SHO','SV','IPouts','ER','BB','SO','ERA']]
boolean_series = pitching2_df.playerID.isin(hof_player_list)
hof_pitching_df = pitching2_df[boolean_series]
hof_pitching_df.head()
```

```python
# Combine all pitchers stats to generate career stats
#ruth_df = hof_batting_df.where(hof_batting_df['playerID']=='ruthba01').dropna()
#career_batting_df = ruth_df[['playerID','G','AB','R','H','2B','3B','HR','RBI','SB']].groupby(['playerID']).sum()
career_pitching_df = hof_pitching_df[['playerID','W','L','GS','CG','SHO','SV','IPouts','ER','BB','SO']].groupby(['playerID']).sum().reset_index()

career_pitching_df
```
